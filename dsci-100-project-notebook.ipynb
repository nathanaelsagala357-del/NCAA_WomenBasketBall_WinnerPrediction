For this classification task — predicting whether a team will win the championship — KNN classification is the appropriate method. Since the outcome variable champion only have to results (Yes/No), KNN works by looking at the K most similar teams based on seed, total_wins_pct, and conf_wins_pct and predicting the majority class among those neighbors.

Why appropriate: KNN is a straightforward classification method that directly maps to our research question. Similar to predicting flower species from petal measurements, we are predicting a category (champion or not) from numeric measurements (seed and win percentages). KNN is intuitive and requires no assumptions about the shape of the data.

Limitations: The main weakness is class imbalance — since very few teams actually win the championship each year, the model may simply predict No for every team and still appear accurate. KNN is also sensitive to the scale of the predictors, so normalizing the variables before fitting is important.

Data processing: The data will be split into 70% training and 30% testing. The training set will be used to fit the model and the test set will be used to evaluate performance. A stratified split on the champion variable will be used to ensure both sets contain a fair proportion of champions and non-champions. No cross validation will be used given the small size of the dataset.

Performance metrics: The model will be assessed using a confusion matrix, which shows how many champions and non-champions were correctly and incorrectly predicted, giving a clearer picture of model performance than accuracy alone.

```r
set.seed(123)

# First is to create a model and then training it and testing it with the actual data
# The training and testing will be using 70 and 30 percent of the actual data respectively
champ_split <- initial_split(champ_data, prop = 0.7, strata = champion)
champ_training <- training(champ_split)
champ_testing <- testing(champ_split)

# Then create cross validation from training data using 5 folds
champ_vfold <- vfold_cv(champ_training, v = 5, strata = champion)

# Then create the recipe for preprocessing
champ_recipe <- recipe(champion ~ seed + total_wins_pct + conf_wins_pct, data = champ_data) |>
step_scale(all_predictors()) |>
step_center(all_predictors())

# Then create the classification model with k-nearest neighbors
champ_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = tune()) |>
set_engine("kknn") |>
set_mode("classification")

# Then create a grid of neighbor values to try 
k_grid <- tibble(neighbors = seq(1, 20, by = 1))

# Then combine recipe and model into a workflow
champ_wf <- workflow() |>
  add_recipe(champ_recipe) |>
  add_model(champ_spec)

# Then find the best k neighbor with the optimal result from the cross validation 
# from 1 to 20 neighbors
best_k <- champ_wf |> 
  tune_grid(resamples = champ_vfold, grid = k_grid) |>
  collect_metrics() |>
  filter(.metric == "accuracy") |>
  arrange(desc(mean)) |>
  slice_head(n = 1) |>
  select(neighbors) |>
  pull()

# Then specify a new model using the best_k
knn_spec_final <- nearest_neighbor(neighbors = best_k) |>
  set_engine("kknn") |>
  set_mode("classification")

# Afterward create a new workflow and fit with the final model
final_fit <- workflow() |>
  add_recipe(champ_recipe) |>
  add_model(knn_spec_final) |>
  fit(data = champ_train)

# Then make predictions on the test data
champ_preds <- final_fit |>
  predict(champ_testing) |>
  bind_cols(champ_testing)

# Check the confusion matrix and accuracy of the algorythm
champ_preds |>
  conf_mat(truth = champion, estimate = .pred_class)

champ_preds |>
  metrics(truth = champion, estimate = .pred_class)
```
